```python
parties=pd.read_csv('Election data 2018.csv')
parties
```

```python
def resolveParty(x):
    max=0
    match=''
    for i in parties['Party']:
        r=f.ratio(x,i)
        if r>max:
            max=r
            match=i
    return match
```

```python
extracted_parties=eledata.Party.apply(resolveParty)
extracted_parties
```

```python
eledata.Party=extracted_parties
```

```python
seats=eledata.groupby(['Party'])['Seat'].count()
```

```python
party_position=seats.nlargest(11)
```
